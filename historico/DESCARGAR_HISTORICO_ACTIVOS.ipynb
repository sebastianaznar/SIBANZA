```python
import yfinance as yf
import pandas as pd 

tickers = [
    "BTC-USD", "ETH-USD", "SOL-USD", "LINK-USD",
    "GC=F", "SI=F", "BZ=F", "NG=F", "HG=F", "ZC=F", "KC=F", "PA=F",
    "TLT", "IEF", "SHY", "TIP", "BNDX", "EMB", "VTC", "JNK", "IBGL.L", "BTP.MI", "MUB",
    "AAPL", "MSFT", "AMZN", "JNJ", "JPM", "XOM", "PG", "TSLA", "UNH", "BRK-B",
    "SAN.MC", "ITX.MC", "IBE.MC", "MC.PA", "SAP.DE", "ASML.AS", "SIE.DE", "NESN.SW",
    "AZN.L", "HSBA.L",
    "2330.TW", "7203.T", "BABA", "TCEHY", "RELIANCE.NS", "VALE", "BHP"
]

data = yf.download(
    tickers,
    period="max",
    group_by="ticker",
    auto_adjust=False
)

rows = []

for ticker in tickers:
    if ticker not in data.columns.levels[0]:
        continue

    df_ticker = data[ticker].copy()
    df_ticker["date"] = df_ticker.index
    df_ticker["ticker"] = ticker

    rows.append(df_ticker.reset_index(drop=True))

final_df = pd.concat(rows, ignore_index=True)

final_df = final_df.dropna(
    subset=["Open", "High", "Low", "Close"]
)

final_df["asset_id"] = (
    final_df["ticker"] + "_" + final_df["date"].astype(str)
)

final_df = final_df[
    ["asset_id", "ticker", "date", "Open", "High", "Low", "Close", "Volume"]
].sort_values(["ticker", "date"])


final_df.to_csv("historical_assets.csv", index=False)
```
