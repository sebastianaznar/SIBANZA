```python
import os
import psycopg2
from dotenv import load_dotenv

# Load environment variables
load_dotenv(override=True)

# Database credentials
USER = os.getenv("user")
PASSWORD = os.getenv("password")
HOST = os.getenv("host")
PORT = os.getenv("port")
DBNAME = os.getenv("dbname")

conn = None

try:
    # 1. Establish connection
    print(f"Connecting to {HOST}...")
    conn = psycopg2.connect(
        user=USER,
        password=PASSWORD,
        host=HOST,
        port=PORT,
        dbname=DBNAME
    )

    cur = conn.cursor()

    # 2. Define and execute price history table creation
    print("Creating 'price_history' table...")
    cur.execute("""
    CREATE TABLE IF NOT EXISTS public.price_history (
        ticker TEXT NOT NULL,
        date DATE NOT NULL,
        open_price DECIMAL(16, 4),
        high_price DECIMAL(16, 4),
        low_price DECIMAL(16, 4),
        close_price DECIMAL(16, 4),
        volume BIGINT,
        PRIMARY KEY (ticker, date),
        CONSTRAINT fk_assets 
            FOREIGN KEY(ticker) 
            REFERENCES public.assets(ticker) 
            ON DELETE CASCADE);
    """)

    # 3. Commit changes
    conn.commit()
    print("Price history table created successfully.")

except Exception as e:
    print(f"Database error: {e}")

finally:
    # 4. Resource cleanup
    if conn:
        cur.close()
        conn.close()
        print("Connection closed.")
```
